```python
numerador = float(input('Entre com o numerador: '))
denomimnador = float(input('Entre com o dominador: '))

chute = float(input ('Entre com seu chute: '))

resultado = numerador/denomimnador

maior = abs(resultado)

#verifica
if abs ((chute - resultado)/maior) <=0.1:
    print('Parabéns!')
else:
    print('O valor correto é {}'.format(resultado))
```
